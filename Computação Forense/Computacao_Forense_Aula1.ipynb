```python
#https://stackoverflow.com/questions/40269605/how-to-create-a-brute-force-password-cracker-for-alphabetical-and-alphanumerical

import itertools
import string
import timeit
from datetime import datetime


def descobrir_senha(real = 'abc', n = 10):
    start = datetime.now()
    characters = string.ascii_lowercase + string.ascii_uppercase + string.digits + string.punctuation
    print('Testando com os caracteres: {}'.format(characters))
    tries = 0
    for length in range(1, n + 1): # quanto maior o range mais possibilidades cria, mais demora
        for guess in itertools.product(characters, repeat = length): # https://docs.python.org/3/library/itertools.html
            tries += 1
            word = ''.join(guess)
            if word == real:
                time = datetime.now() - start
                return print('ANSWER: Match: {} , Tries: {}, Time: {}'.format(word, tries, time))

    time = datetime.now() - start
    return print('ANSWER: No match, Time: {}'.format(time))
```

```python
descobrir_senha('jo1ha', 10)
```

```python
#https://wiki.skullsecurity.org/index.php/Passwords

#http://downloads.skullsecurity.org/passwords/rockyou.txt.bz2

from bz2 import BZ2File as bzopen

with bzopen("/content/rockyou.txt.bz2", "r") as bzfin:
    """ Handle lines here """
    lst_senhas = []
    for i, line in enumerate(bzfin):
        lst_senhas.append(line.strip().decode("utf-8", errors = 'ignore'))
```

```python
def lista_senha(lst, real = 'abc'):
    start = datetime.now()
    tries = 0
    for senha in lst:
        tries += 1
        if senha == real:
            time = datetime.now() - start
            return print('ANSWER: Match: {} , Tries: {}, Time: {}'.format(senha, tries, time))
    
    time = datetime.now() - start
    return print('ANSWER: No match, Tries: {}, Time: {}'.format(tries, time))

```

```python
lista_senha(lst_senhas, 'jo1ha')
```

```python
lista_senha(lst_senhas, 'naruto123')
```

```python
def lista_senha_v1(lst, real = 'abc'):
    start = datetime.now()
    tries = 0
    for senha in lst:
        for combinacao in list(map(''.join, itertools.product(*zip(senha.upper(), senha.lower())))):
            tries += 1
            if combinacao == real:
                time = datetime.now() - start
                return print('ANSWER: Match: {} , Tries: {}, Time: {}'.format(combinacao, tries, time))
    
    time = datetime.now() - start
    return print('ANSWER: No match, Tries: {}, Time: {}'.format(tries, time))
```
