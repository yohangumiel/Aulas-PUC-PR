```python
geneSet = " abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ!.@#$%&1234567890"
target = "senha nova123@"
```

```python
import random

def generate_parent(length):
    genes = []
    while len(genes) < length:
        sampleSize = min(length - len(genes), len(geneSet))
        genes.extend(random.sample(geneSet, sampleSize))
    return ''.join(genes)

import random
def generate_parent1(length):
    genes = []
    while len(genes) < length:
        print('length: ', length)
        print('genes: ', len(genes))
        sampleSize = min(length - len(genes), len(geneSet))
        print('sample size: ', sampleSize)
        genes.extend(random.sample(geneSet, sampleSize))
    return ''.join(genes)
```

```python
def get_fitness(guess):
    return sum(1 for expected, actual in zip(target, guess) if expected == actual)

def get_fitness1(guess):
    fitness = 0
    for expected, actual in zip(target,guess):
        print(expected, actual)
        if expected == actual:
            print('match: ', expected, actual)
            fitness+=1
    return fitness
```

```python
def mutate(parent):
    index = random.randrange(0, len(parent)) # selecionando um index aleatório da string
    childGenes = list(parent)
    newGene, alternate = random.sample(geneSet, 2)
    if childGenes[index] == newGene:
        childGenes[index] = alternate
    else:
        childGenes[index] = newGene

    return ''.join(childGenes)

def mutate1(parent):
    index = random.randrange(0, len(parent)) # selecionando um index aleatório da string
    print(index)
    childGenes = list(parent)
    print(childGenes)
    newGene, alternate = random.sample(geneSet, 2)
    print('novos possíveis genes: ', newGene, alternate)
    if childGenes[index] == newGene:
        childGenes[index] = alternate
    else:
        childGenes[index] = newGene

    return ''.join(childGenes)
```

```python
import datetime

def display(guess):
    timeDiff = datetime.datetime.now() - startTime
    fitness = get_fitness(guess)
    print("\n{0}\t{1}\t{2}".format(guess, fitness, str(timeDiff)))
```

```python
random.seed()
startTime = datetime.datetime.now()
bestParent = generate_parent(len(target))
bestFitness = get_fitness(bestParent)
display(bestParent)


while True:
    child = mutate(bestParent)
    childFitness = get_fitness(child)
    display(child)
    if bestFitness >= childFitness:
        continue
    # display(child)
    if childFitness >= len(bestParent):
        break
    bestFitness = childFitness
    bestParent = child
```
